```python
BAD_FEATURES= [
    'Empresa',
    'Setor',
    'Sede (UF)',
    'Posição',
    'ID',
    'Pontuação',
]

BAD_FEATURES = np.append(BAD_FEATURES, ranking.columns[ranking.columns.str.contains('Ranking')])

FEATURES = ranking.columns.difference(BAD_FEATURES)

X = ranking[FEATURES]

for i in X.columns:
    X[i] = pd.to_numeric(X[i], errors='coerce').fillna(0)

# X[X.columns] = preprocessing.StandardScaler().fit_transform(X)

X.describe().transpose()
```

```python
data = []
for i in range(2,10):
    cm = cluster.KMeans(n_clusters=i, random_state=42).fit(X)
    ac = cluster.AgglomerativeClustering(n_clusters=i, linkage='ward').fit(X)
    data += [[
         i,
         metrics.silhouette_score(X, cm.labels_, metric='euclidean'),
         metrics.silhouette_score(X, ac.labels_, metric='euclidean'),
        ]]

pd.DataFrame(data, columns=['clusters', 'kmeans', 'hierarchical'])
```
